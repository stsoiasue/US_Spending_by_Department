```python
import json
import requests
import pandas as pd
import numpy as np

import sqlite3
```

```python
conn = sqlite3.connect('us_data3.sqlite')
cnx = sqlite3.connect('final_dataset.sqlite')

us_data_df = pd.read_sql_query("SELECT * FROM department_contracts", conn)
zip_code_df = pd.read_sql_query("SELECT * FROM department_contracts", cnx)
zip_code_df = zip_code_df.drop('index', 1)
us_data_df = us_data_df.drop('index', 1)

us_data_df.head()
```

```python
zip_code_df.head()
```

```python
merged_df = us_data_df.merge(zip_code_df, how='left', on='POP_Zip')
merged_df = merged_df[np.isfinite(merged_df['Latitude'])]
merged_df.head()
```

```python
conn = sqlite3.connect('end_of_class.sqlite')

merged_df.to_sql("department_contracts", conn, if_exists="replace")
merged_df.to_csv('end_of_class.csv')
```
