```python
import json
import requests
import pandas as pd
import numpy as np

import sqlite3
```

```python
conn = sqlite3.connect('us_data3.sqlite')
cnx = sqlite3.connect('final_dataset.sqlite')

us_data_df = pd.read_sql_query("SELECT * FROM department_contracts", conn)
zip_code_df = pd.read_sql_query("SELECT * FROM department_contracts", cnx)
zip_code_df = zip_code_df.drop('index', 1)
us_data_df = us_data_df.drop('index', 1)

us_data_df.head()
```

```python
zip_code_df.head()
```

```python
merged_df = us_data_df.merge(zip_code_df, how='left', on='POP_Zip')
merged_df = merged_df[np.isfinite(merged_df['Latitude'])]
merged_df.head()
```

```python
dot_df = merged_df.loc[merged_df['Awarding_Agency'] == 'Department of Transportation']
dot_df['Total_Obligation'] = dot_df['Total_Obligation'].apply(pd.to_numeric)
dot_df = dot_df.sort_values(by=['Total_Obligation'], ascending=False)
dot_df = dot_df.head(10)
dot_df
```

```python
doa_df = merged_df.loc[merged_df['Awarding_Agency'] == 'Department of Agriculture']
doa_df['Total_Obligation'] = doa_df['Total_Obligation'].apply(pd.to_numeric)
doa_df = doa_df.sort_values(by=['Total_Obligation'], ascending=False)
doa_df = doa_df.head(10)
doa_df
```

```python
doi_df = merged_df.loc[merged_df['Awarding_Agency'] == 'Department of Interior']
doi_df['Total_Obligation'] = doi_df['Total_Obligation'].apply(pd.to_numeric)
doi_df = doi_df.sort_values(by=['Total_Obligation'], ascending=False)
doi_df = doi_df.head(10)
doi_df
```
