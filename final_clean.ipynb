```python
import json
import requests
import pandas as pd
import numpy as np

import sqlite3
```

```python
conn = sqlite3.connect('data.sqlite')
us_data_df = pd.read_sql_query("SELECT * FROM department_contracts", conn)
us_data_df = us_data_df.drop(['level_0', 'index'], axis=1)
merged_df = merged_df[np.isfinite(merged_df['Latitude'])]
us_data_df.head()
```

```python
doa_df = us_data_df.loc[us_data_df['Awarding_Agency'] == 'Department of Agriculture']
doa_df['Total_Obligation'] = doa_df['Total_Obligation'].apply(pd.to_numeric)
doa_df = doa_df.sort_values(by=['Total_Obligation'], ascending=False)
doa_df = doa_df.head(10)
doa_df
```

```python
dot_df = us_data_df.loc[us_data_df['Awarding_Agency'] == 'Department of Transportation']
dot_df['Total_Obligation'] = dot_df['Total_Obligation'].apply(pd.to_numeric)
dot_df = dot_df.sort_values(by=['Total_Obligation'], ascending=False)
dot_df = dot_df.head(10)
dot_df
```

```python
doi_df = us_data_df.loc[us_data_df['Awarding_Agency'] == 'Department of the Interior']
doi_df['Total_Obligation'] = doi_df['Total_Obligation'].apply(pd.to_numeric)
doi_df = doi_df.sort_values(by=['Total_Obligation'], ascending=False)
doi_df = doi_df.head(10)
doi_df
```

```python
top = [doi_df, doa_df, dot_df]
top_ten_df = pd.concat(top)
# top_ten_df = top_ten_df.drop(['index', 'level_0'], axis=1)


top_ten_df
```

```python
conn = sqlite3.connect('data.sqlite')

top_ten_df.to_sql("top_ten", conn, if_exists="replace")
us_data_df.to_sql("lat_lon", conn, if_exists="replace")
```
