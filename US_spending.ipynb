```python
import json
import requests
import pandas as pd
import numpy as np

import sqlite3
```

```python
page = 1
url = 'https://api.usaspending.gov/api/v1/awards/?limit=500&page=' + str(page)



params = {
    "filters": [
      {
        "field": "date_signed",
        "operation": "greater_than_or_equal",
        "value": "2017-01-01"
      },
      {
        "field": "date_signed",
        "operation": "less_than",
        "value": "2017-12-31"
      }, 
      {
        "field": "category",
        "operation": "equals",
        "value": "contract"
      },
      {
        "field": "awarding_agency__toptier_agency__cgac_code",
        "operation": "in",
        "value": ["012", "069", "014"]
      },
        {
        "field": "place_of_performance__country_name",
        "operation": "equals",
        "value": "UNITED STATES"
        }]
}
```

```python
us_data = requests.get(url , json =params).json()
print(json.dumps(us_data, indent=4, sort_keys=True))
```
